### How much caffeine should I take to have a steady amount in my system?
Assuming a "budget" of 400mg a day.

```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
```

```python
MINS_PER_HOUR = 60
HOURS_PER_DAY = 24
XLIM = HOURS_PER_DAY * MINS_PER_HOUR
HALF_LIFE = 5.5 * MINS_PER_HOUR
minutes_in_day = np.arange(0, XLIM)

def take(hour):
    minutes = hour * MINS_PER_HOUR
    caffeine_curve = 1 / (2 ** (minutes_in_day / HALF_LIFE)) * 100
    left_padding = np.repeat(0, minutes)
    caffeine = np.concatenate((left_padding, caffeine_curve))
    return caffeine[:XLIM]

def plot(x, y):
    fig, ax = plt.subplots()
    axes = plt.gca()
    axes.set_xlim([0, XLIM])
    axes.set_xlabel("minutes after having caffeine")
    axes.set_ylim([0, 250])
    axes.set_ylabel("caffeine (mg)")
    line, = ax.plot(x, y, linewidth=2)
    plt.show()
```

```python
total_caffeine = np.sum((take(0), take(0), take(4), take(8)), axis=0)
print("Caffeine in your system 4 hours into your day:", total_caffeine[4 * MINS_PER_HOUR])
print("Caffeine in your system 8 hours into your day:", total_caffeine[8 * MINS_PER_HOUR])
print("Caffeine in your system 12 hours into your day:", total_caffeine[12 * MINS_PER_HOUR])
print("Caffeine in your system at bedtime:", total_caffeine[16 * MINS_PER_HOUR])
plot(minutes_in_day, total_caffeine)
```
